```python
# Imports

! pip install opencv-python
import cv2
import os
from IPython.display import Image 
```

```python
path = '/content/'
image = cv2.imread(path + 'bird3.jpg')
Image('/content/bird3.jpg')
```

```python
# 648x454
# 190-326  323-493
# 112-338  113-340  

# draw figure
x1, y1 = 190, 103
x2, y2 = 328, 329

x3, y3 = 323, 113
x4, y4 = 493, 340

color = [238, 151, 255]
width = 2

image_with_rectangle = cv2.rectangle(image, (x1, y1), (x2, y2), color, width)
image_with_rectangles = cv2.rectangle(image, (x3, y3), (x4, y4), color, width)
cv2.imwrite('image_with_rectangles.jpg', image_with_rectangles)
Image('image_with_rectangles.jpg')
```

```python
# draw text
x1, y1 = 200, 100

x3, y3 = 333, 110

font_size = 1
font = cv2.FONT_HERSHEY_SIMPLEX
width = 2
color = [238, 151, 255]
text = 'bird'

image_with_text = cv2.putText(image_with_rectangles, text, (x1, y1), font, font_size, color, width)
image_with_texts = cv2.putText(image_with_text, text, (x3, y3), font, font_size, color, width)
cv2.imwrite('image_with_texts.jpg', image_with_texts)
Image('image_with_texts.jpg')
```

```python
# flip image

flip_image = cv2.flip(image,1)
cv2.imwrite('flip_image.jpg', flip_image)
Image('flip_image.jpg')
```

```python
# sepia
import random
from PIL import Image, ImageDraw 
import matplotlib.pyplot as plt

mode = 1
image = Image.open('flip_image.jpg') 
draw = ImageDraw.Draw(image)  
width = image.size[0]
height = image.size[1]	
pix = image.load()

if (mode == 1):
	depth = 30
	for i in range(width):
		for j in range(height):
			a = pix[i, j][0]
			b = pix[i, j][1]
			c = pix[i, j][2]
			S = (a + b + c) // 3
			a = S + depth * 2
			b = S + depth
			c = S
			if (a > 255):
				a = 255
			if (b > 255):
				b = 255
			if (c > 255):
				c = 255
			draw.point((i, j), (a, b, c))

image.save('sepia.jpg', 'JPEG')
del draw

im = Image.open('sepia.jpg')
plt.imshow(im)
plt.show()
```
