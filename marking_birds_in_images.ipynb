```python
# Imports

! pip install opencv-python
import cv2
import os
from IPython.display import Image 
```

```python
path = '/content/'
image = cv2.imread(path + 'bird3.jpg')
Image('/content/birds.jpg')
```

```python
# 648x454
# 190-326  323-493
# 112-338  113-340  

# draw figure
x1, y1 = 190, 103
x2, y2 = 328, 329

x3, y3 = 323, 113
x4, y4 = 493, 340

color = [238, 151, 255]
width = 2

image_with_rectangle = cv2.rectangle(image, (x1, y1), (x2, y2), color, width)
image_with_rectangles = cv2.rectangle(image, (x3, y3), (x4, y4), color, width)
cv2.imwrite('image_with_rectangles.jpg', image_with_rectangles)
Image('image_with_rectangles.jpg')
```

```python
# draw text
x1, y1 = 200, 100

x3, y3 = 333, 110

font_size = 1
font = cv2.FONT_HERSHEY_SIMPLEX
width = 2
color = [238, 151, 255]

text = 'bird'

image_with_text = cv2.putText(image_with_rectangles, text, (x1, y1), font, font_size, color, width)

image_with_texts = cv2.putText(image_with_text, text, (x3, y3), font, font_size, color, width)
cv2.imwrite('image_with_texts.jpg', image_with_texts)
Image('image_with_texts.jpg')
```

```python
os.remove('image_with_rectangles.jpg')
del(image_with_rectangle)

os.remove('image_with_texts.jpg')
del(image_with_text)
```
